```python
df = pd.concat([df1, df2, df3, df4, df5, df6, df7, df8, df9, df10, df11, df12])
```

```python
df.head()
```

```python
df.describe()
```

```python
df['ended_at'] = pd.to_datetime(df['ended_at'])
df['started_at'] = pd.to_datetime(df['started_at'])
df['day_of_week'] = df['started_at'].dt.day_name()
```

```python
df['ride_length'] = (df['ended_at'] - df['started_at']).dt.total_seconds()/60
```

```python
df.head()
```

```python
df.isnull().sum()
```

# Removing Null values from start_station_name, start_station_id and end_station_id

```python
df = df[df['end_station_id'].notna()]
df = df[df['start_station_id'].notna()]
df = df[df['start_station_name'].notna()]
```

```python
df.head()
```

```python
df.notna().sum()
```

```python
df['day_of_week_number']=df['day_of_week']
df['day_of_week_number'].replace(['Sunday','Monday','Tuesday','Wednesday','Thursday','Friday','Saturday'] ,[1,2,3,4,5,6,7],inplace=True)
df.head()
```

# Ride length for different types of members

```python
plt.figure(figsize = (10, 10))
sns.barplot(x = df["member_casual"], y = df["ride_length"], palette = "Set2");
```

```python
sorted_df=df.sort_values(by="day_of_week_number")
plt.figure(figsize = (10, 10))
sns.barplot(x = sorted_df["day_of_week_number"], y = sorted_df["ride_length"],hue=sorted_df["member_casual"], palette = "Set2");
```

# Differentiating data based on bike preference by different types of members

```python
df['Ride_values']=df['rideable_type']
df['Ride_values'].replace(['electric_bike','classic_bike','docked_bike'] ,[1,1,1],inplace=True)
```

```python
df.head()
```

```python
ty=df[["rideable_type","member_casual","Ride_values"]].groupby(["rideable_type","member_casual"],as_index=False).sum("Ride_values")
ty.head()
```

```python
plt.figure(figsize = (10, 10))
sns.barplot(x = ty["member_casual"], y = ty["Ride_values"],hue =ty["rideable_type"], palette = "Set2");
```

# Stations with most outcoming and incoming rides being used by Casual Members

```python
u=['casual']
df2=df[df.member_casual.isin(u)]
```

```python
df2.head()
```

```python
plt.figure(figsize = (20,8))
ssnc = df2["start_station_name"].value_counts()[:15].plot(kind = 'bar', color = "gray")
for p in ssnc.patches:
    ssnc.annotate(int(p.get_height()), (p.get_x() + 0.25, p.get_height() + 1), ha = 'center', va = 'bottom', color = 'black')
```

```python
plt.figure(figsize = (20,8))
esnc = df2["end_station_name"].value_counts()[:15].plot(kind = 'bar', color = "gray")
for p in esnc.patches:
    esnc.annotate(int(p.get_height()), (p.get_x() + 0.25, p.get_height() + 1), ha = 'center', va = 'bottom', color = 'black')
```
